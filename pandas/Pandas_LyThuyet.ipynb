## Data cleaning basics

```
data_copy = data.copy()
```

**Modifying columns in a dataframe**

```
# Renaming an existing column
data_copy.rename(columns={'Toan': 'Math'}, inplace=True)
data_copy.head()
```

```
# Modifying a string column
data_copy['Ma_mon_ngoai_ngu'] = data_copy['Ma_mon_ngoai_ngu'].str.replace('N', 'M')
data_copy.head()
```

```
# Convert type column
data_copy['sbd'] = data_copy['sbd'].astype(str)
data_copy['sbd'][0]
```

**String column operations**

```
# Extracting values from strings
data_copy['Van'] = data_copy['Van'].astype(str).str.split('.').str[0]
data_copy.head()
```

# Data Aggregation

```
data_copy = data.copy()
```

**Groupby Object**

```
# Create a groupby object and apply a function to that
data_copy.groupby('Ma_mon_ngoai_ngu').agg('mean')
```

```
# Apply multiple functions to a GroupBy object
data_copy.groupby('Ma_mon_ngoai_ngu').agg(['mean', 'sum'])
```

# Transforming data 

```
# Apply a function element-wise to a series
def convert_sbd(x):
    x = str(x)
    if len(x) == 7:
        x = '0' + x
    return x[:2]

data_copy['ma_vung'] = data_copy['sbd'].apply(lambda x: convert_sbd(x))
```
